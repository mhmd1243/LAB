```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Forward bias region
V_forward = np.linspace(0, 1, 100)
I_forward = np.exp(40 * (V_forward - 0.7)) - 1  # Exponential growth after threshold
I_forward[V_forward < 0.7] = 0  # No current before threshold

# Reverse bias region
V_reverse = np.linspace(-1, 0, 50)
I_reverse = -1e-6 * np.ones_like(V_reverse)  # Small leakage current
I_reverse[V_reverse < -0.9] = -5e-3  # Breakdown region

# Combine data
V = np.concatenate([V_reverse, V_forward])
I = np.concatenate([I_reverse, I_forward])

# Create a Seaborn line plot
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
sns.lineplot(x=V, y=I, label='V-I Characteristic', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.title('V-I Characteristic Curve of a Silicon Diode')
plt.legend()
plt.show()
```
