```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate voltage values for forward and reverse bias
V_forward = np.linspace(0, 1, 100)
I_forward = np.exp(40 * (V_forward - 0.7)) - 1  # Exponential growth after threshold
V_reverse = np.linspace(-1, 0, 100)
I_reverse = -0.0001 * np.ones_like(V_reverse)  # Small leakage current

# Combine forward and reverse bias regions
V = np.concatenate((V_reverse, V_forward))
I = np.concatenate((I_reverse, I_forward))

# Plot using seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(8, 6))
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.plot(V, I, label='Diode V-I Characteristic', color='blue')

# Labels
plt.text(-0.8, -0.0002, "Reverse Bias", fontsize=12, color='red')
plt.text(0.4, 0.5, "Forward Bias", fontsize=12, color='green')
plt.xlabel("Voltage (V)")
plt.ylabel("Current (I)")
plt.title("V-I Characteristic Curve of a Silicon Diode")
plt.legend()
plt.show()
```
